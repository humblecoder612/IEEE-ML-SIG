```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X,Y = make_classification(n_samples=100,n_clusters_per_class=1,n_features=2,n_classes=2,n_informative=2,n_redundant=0,random_state = 1)
```

```python
plt.scatter(X[:,0],X[:,1],c = Y)
```

```python
def dist(x1,x2):
    return np.sum((x2 - x1)**2)**.5
```

```python
def KNN(X,Y,qp,k = 5):
    m = X.shape[0]
    vals = []
    for i in range(m):
        d = dist(X[i],qp)
        vals.append((d,Y[i]))
    
    new_vals = np.array(sorted(vals))[:,1]
    new_vals = new_vals[:k]
    t =  np.unique(new_vals,return_counts=True)
    idx = np.argmax(t[1])
    pred = int(t[0][idx])
    return pred
```

```python
print(KNN(X,Y,querypoint1))
print(KNN(X,Y,querypoint2))
```
