```python
## Importing Libraries.

from sklearn.linear_model import LinearRegression # From sklearn library in linear model get Linear Regression.  
import numpy as np # Numpy Is a Library used for Array.
import matplotlib.pyplot as plt #  Matplotlib is used for Ploting Graph.
```

```python
## Dataset

Hours_of_study = np.array([ 3, 4, 5, 6, 4, 9, 7]) # Hours a student studies.
percen_student = np.array([85,89,91,95,92,98,93]) # Percantage He/She Achieved
```

```python
## PLOT

plt.scatter(Hours_of_study, percen_student)
plt.show()
```

```python
# TRAINING DATA

regr = LinearRegression() ## Calling Linear Regression
Hours_of_study_reshaped = Hours_of_study.reshape(Hours_of_study.shape[0],1) # Reshaping Data -> row matrix to coloumn matrix
regr.fit(Hours_of_study_reshaped,percen_student) # Training Data.
```

```python
# PREDICTION

xp = [[0],[4],[10]] # TEST CASE 0HRS, 4Hrs, 10 Hrs
yp = regr.predict(xp) # Predicting Value

# PLOTING THE LINE.
%matplotlib inline
plt.plot(xp, yp, marker='*', Color="b")
plt.plot(Hours_of_study,percen_student,  'g^')
plt.show()
```
